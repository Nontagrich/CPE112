## Bubble Sort 

```python
def bubbleSort(seq):
    n = len(seq) - 1
    for i in range(n,0,-1):
        for j in range(i):
            if seq[j] > seq[j+1]:
                seq[j], seq[j+1] = seq[j+1], seq[j]
                #print(seq)
    return seq
```

## Selection Sort 

```python
def selectionSort(seq):
    n = len(seq)
    for i in range(n-1):
        smallNdx = i
        for j in range(i+1, n):
            if seq[j] < seq[smallNdx]:
                smallNdx = j
        if smallNdx != i:
            seq[i], seq[smallNdx] = seq[smallNdx], seq[i]
```

> test and learn selection sort algorithm

```python
n = 20
lum = [random.randint(1,100) for x in range(n)]
lum
```

```python
selectionSort(lum)
lum
```

## Insertion Sort

```python
def insertionSort(seq):
    n = len(seq)
    for i in range(1, n):
        value = seq[i]
        pos = i
        while pos > 0 and value < seq[pos - 1]:
            seq[pos] = seq[pos - 1]
            pos -= 1
        seq[pos] = value
    return seq
```

> test and learn insertion sort algorithm

```python
n = 300
lum = [random.randint(1,100) for x in range(n)]
lum
```

```python
insertionSort(lum)
lum
```
